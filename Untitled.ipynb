```python
import pandas as pd
import matplotlib.pyplot as plt

f = pd.read_csv('Blood_samples_dataset_balanced_2(f).csv')

# Extract columns for Glucose and Hemoglobin
glucose = f.iloc[:, 0]  # First column for Glucose
hemoglobin = f.iloc[:, 2]  # Third column for Hemoglobin

# Scatter plot for Glucose vs Hemoglobin
plt.figure(figsize=(8, 6))
plt.scatter(glucose, hemoglobin, color='blue', alpha=0.5)
plt.title('Glucose vs Hemoglobin')
plt.xlabel('Glucose')
plt.ylabel('Hemoglobin')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

f = pd.read_csv('Blood_samples_dataset_balanced_2(f).csv')

disease = f.iloc[:, -1]
cholesterol = f.iloc[:, 1]
hemoglobin = f.iloc[:, 2]

data = pd.DataFrame({'Disease': disease, 'Cholesterol': cholesterol, 'Hemoglobin': hemoglobin})

plt.figure(figsize=(10, 8))
sns.scatterplot(x='Disease', y='Cholesterol', data=data, color='blue', alpha=0.5, label='Cholesterol')
sns.scatterplot(x='Disease', y='Hemoglobin', data=data, color='red', alpha=0.5, label='Hemoglobin')
plt.title('Cholesterol and Hemoglobin by Disease')
plt.xlabel('Disease')
plt.ylabel('Level')
plt.legend()
plt.grid(True)
plt.show()
```
